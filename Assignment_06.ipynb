21.  Why is NumPy’s array slicing faster than Python’s list slicing ?

   - NumPys array slicing is faster than Pythons list slicing because NumPy stores data in one continuous block of memory while Python lists store elements separately with pointers. This makes NumPy more efficient—when you select an array  it doesn’t create a new copy it just provides a view of the original data, so changes happen in the same memory space. But with Python lists, slicing creates a whole new list  which takes extra time and memory. NumPy is also written in C so it runs much faster than Pythons built-in lists. That’s why NumPy is a top choice for handling large datasets in data science and machine learning .
